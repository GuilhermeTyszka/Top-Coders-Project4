### Customer Personality Analysis
#### About Dataset
##### Problem Statement
<font size="2">Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.  
Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.</font>

##### Content

<font size="2">

**People**:   
**ID:** Customer's unique identifier  
**Year_Birth:** Customer's birth year  
**Education:** Customer's education level  
**Marital_Status:** Customer's marital status  
**Income:** Customer's yearly household income  
**Kidhome:** Number of children in customer's household  
**Teenhome:** Number of teenagers in customer's household  
**Dt_Customer:** Date of customer's enrollment with the company  
**Recency:** Number of days since customer's last purchase  
**Complain:** 1 if the customer complained in the last 2 years, 0 otherwise 

**Products**  
**MntWines:** Amount spent on wine in last 2 years  
**MntFruits:** Amount spent on fruits in last 2 years  
**MntMeatProducts:** Amount spent on meat in last 2 years  
**MntFishProducts:** Amount spent on fish in last 2 years  
**MntSweetProducts:** Amount spent on sweets in last 2 years  
**MntGoldProds:** Amount spent on gold in last 2 years  

**Promotion**  
**NumDealsPurchases:** Number of purchases made with a discount  
**AcceptedCmp1:** 1 if customer accepted the offer in the 1st campaign, 0 otherwise  
**AcceptedCmp2:** 1 if customer accepted the offer in the 2nd campaign, 0 otherwise  
**AcceptedCmp3:** 1 if customer accepted the offer in the 3rd campaign, 0 otherwise  
**AcceptedCmp4:** 1 if customer accepted the offer in the 4th campaign, 0 otherwise  
**AcceptedCmp5:** 1 if customer accepted the offer in the 5th campaign, 0 otherwise  
**Response:** 1 if customer accepted the offer in the last campaign, 0 otherwise  

**Place**  
**NumWebPurchases:** Number of purchases made through the company’s website  
**NumCatalogPurchases:** Number of purchases made using a catalogue  
**NumStorePurchases:** Number of purchases made directly in stores  
**NumWebVisitsMonth:** Number of visits to company’s website in the last month

**Dataset source link**: https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis/discussion</font>

## Limpeza e Ajustes do DataFrame

```python
# Limpeza de algumas colunas

df_customer.drop(['Z_Revenue', 'Z_CostContact'], axis=1)

# Correção dos Nomes
correcao_nomes = {'YOLO':'Single',
            'Alone':'Single',
            'Absurd' :'Single',}

df_customer = df_customer.replace(correcao_nomes)

```

## Criação de Novas Colunas

```python
df_customer['Total_Children'] = df_customer['Kidhome'] + df_customer['Teenhome']
df_customer['MntTotal'] = df_customer['MntWines'] + df_customer['MntFruits'] + df_customer['MntMeatProducts'] + df_customer['MntFishProducts'] + df_customer['MntSweetProducts'] + df_customer['MntGoldProds']
df_customer['Accepted_offer'] = df_customer.AcceptedCmp1 + df_customer.AcceptedCmp2 + df_customer.AcceptedCmp3 + df_customer.AcceptedCmp4 + df_customer.AcceptedCmp5
df_customer['Accepted_offer'] = df_customer.Accepted_offer.apply(lambda x: 0 if x == 0 else 1)
df_customer['Have_sons'] = df_customer.Total_Children.apply(lambda x: 'Não' if x == 0 else 'Sim')
df_customer['Dt_Customer'] = pd.to_datetime(df_customer['Dt_Customer'], format = '%d-%m-%Y')
df_customer['Age'] = 2015 - df_customer['Year_Birth']
```

## Criação de Novas Classificações


```python
max_age = df_customer['Age'].max()
min_age = df_customer['Age'].min()
labels_age = ['Adults', 'Middle Age', 'Elder']
classes_age = [min_age, 40, 60, max_age]
df_customer['Age_group'] = pd.cut(x = df_customer['Age'], bins = classes_age, labels = labels_age, include_lowest = True)

labels_income = ['Low Income', 'Middle Income', 'High Income']
classes_income = [0, 30000, 60000, 1000000]
df_customer['Income_group'] = pd.cut(x = df_customer['Income'], bins = classes_income, labels = labels_income, include_lowest = True)
```

## Realizando agrupamentos para estudo

```python
marital_status_grouping = df_customer.groupby(by = 'Marital_Status')
sons_or_not_grouping = df_customer.groupby(by = 'Have_sons')
age_grouping = df_customer.groupby(by = 'Age_group')
Income_grouping = df_customer.groupby(by = 'Income_group')
Education_grouping = df_customer.groupby(by = 'Education')
```

## Removendo Outliers

```python
def outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_boundary = Q1 - 1.5 * IQR
    upper_boundary = Q3 + 1.5 * IQR
    df_customer = dataframe.loc[(dataframe[column] >= lower_boundary) & (dataframe[column] <= upper_boundary)]
    return df_customer

df_customer = outliers(df_customer, 'Year_Birth')

income_outlier = df_customer[df_customer['Income']==666666.0].index
df_customer.drop(income_outlier, inplace=True)
```

## Qual a relação entre o consumo de cada produto e escolaridade?

```python
# gráfico para entender que tipo de produto é interessante para cada grau de escolaridade dos clientes
columns_Mnt = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df_educ_prod_mean = df_customer.pivot_table(index='Education', values=columns_Mnt, aggfunc='mean')
df_educ_prod_mean.rename(mapper = {'MntWines': 'Vinhos', 'MntFruits': 'Frutas', 'MntMeatProducts': 'Carnes', 'MntFishProducts': 'Peixes', 'MntSweetProducts': 'Doces', 'MntGoldProds':'Ouro'}, inplace=True, axis = 1)
df_educ_prod_mean = df_educ_prod_mean.reindex(['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']).rename({'Graduation': 'Graduado', 'PhD': 'Doutor', 'Master': 'Mestre', '2n Cycle': '2° Ciclo', 'Basic': 'Básico'})
fig = px.bar(df_educ_prod_mean, barmode='group', height=300, width=700)

fig.update_layout(
    xaxis_title=None,
    yaxis_title={
        'text': "Média da quantidade de produtos",
        'font_size': 9
    },
    legend_title={
        'text': "Produtos:",
        'font_size': 12
    },
    template='none',
    title={
        'text': 'Média dos produtos comprados por escolaridade',
        'y':0.95,
        'x':0.25,
        'font_color': 'black',
        'font_size': 15
    }
)
```

```python
# gráfico de pizza para verificar a distribuição do nível de escolaridade do dataset
color=sns.color_palette('colorblind')
to_plot = df_customer.groupby('Education')['ID'].count().rename({'Graduation': 'Graduado', 'PhD': 'Doutor', 'Master': 'Mestre', '2n Cycle': '2° Ciclo', 'Basic': 'Básico'})
plt.pie(to_plot, autopct='%1.0f%%', colors=color)
plt.legend(to_plot.index, loc = 'best', bbox_to_anchor=(0.9, 0., 0.4, 1));
```

### É possível notar que o produto mais comprado é o vinho seguindo pela carne, nota-se também que a compra média de vinho tem um crescimento de acordo com o aumento do grau escolar.

## Local de compra de preferência por grupo de faixa etária

```python
local_per_age_group = age_grouping.mean()[['NumWebPurchases', 'NumDealsPurchases',
                                                       'NumCatalogPurchases', 'NumStorePurchases']].T
for i in local_per_age_group.columns:
    local_per_age_group[i] = local_per_age_group[i]/local_per_age_group[i].sum()

fig, axs = plt.subplots(1, 3, figsize=(12,4), sharey=True)

explode = [0, 0, 0, 0.1] 

axs[0].pie(local_per_age_group['Adults'], autopct='%.1f%%', explode = explode, shadow = True)
axs[0].set(title='Jovens Adultos')

axs[1].pie(local_per_age_group['Middle Age'], autopct='%.1f%%', explode = explode, shadow = True)
axs[1].set(title='Adultos')

axs[2].pie(local_per_age_group['Elder'], autopct='%.1f%%', explode = explode, shadow = True)
axs[2].set(title='Idosos')

plt.legend(local_per_age_group.index, ncol = 4)
```

### Nota-se que loja física é o local preferido de realizar compras pelos clientes, seguido por compras na internet. Porém, não conseguimos observar mudança de preferências entre as distintas faixas etárias.
